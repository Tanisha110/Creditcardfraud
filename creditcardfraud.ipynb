```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv', on_bad_lines='skip', engine='python')
```

```python
credit_card_data
```

```python
credit_card_data.info()
```

```python
credit_card_data['Class'].value_counts()
```

this datais highly unbalanced
0= normal transaction
1= fraudulent transaction

```python
legit =credit_card_data[credit_card_data.Class==0]
fraud =credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare he values for bothe transactions
credit_card_data.groupby('Class').mean()
```

#Under-Sampling
build a sample datase containing similar distribution of normal and fraudulent transactions

```python
legit_sample = legit.sample(n=73) #put n whatever the no of fraud transactions are there in your data
```

##Concatinating two dataset

```python
new_dataset= pd.concat([legit_sample ,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

##splitting the data into targets and frame

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

##Split the training data and testing data

```python
X_train , X_test , Y_train , Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print (X.shape,X_train.shape,X_test.shape)
```

##model training

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

##model evaluation
Accuracy score

```python
X_train_prediction = (model.predict(X_train))
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy on training data :', training_data_accuracy)
```

```python
X_test_prediction = (model.predict(X_test))
training_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on test data :', training_data_accuracy)
```
